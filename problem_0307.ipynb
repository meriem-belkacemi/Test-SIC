```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

#### Answer the following questions by providing Python code:

We will use the dataset "data_ToothGrowth.csv". 
- len = teeth length. 
- supp = supplementary type with two possible values: VC (vitamin C) or OJ (orange juice). 
- dose (mgr/day) = dosage per day with three possible values: 0.5, 1.0, 2.0.

```python
df = pd.read_csv('data_ToothGrowth.csv', header='infer')
```

1). Perform the EDA and ANOVA ($\alpha=$ 5%) with the 'supp' variable. <br>
Summarize the result.

```python
# Categories of 'supp'. 
df.supp.unique()
```

```python
# Group by category.
x_vc = len(df[df.supp=='VC'])
x_oj = len(df[df.supp=='OJ'])
print (x_vc, x_oj)
```

```python
# Multiple Boxplots.
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.boxplot([x_vc,x_oj],0)
plt.show()
```

```python
# ANOVA.
res = st.f_oneway(df.supp=='VC',df.supp=='OJ')

print("Test statictic  : %f " % res[0] )
print("p-value         : %f " % res[1] )
```

2). Perform the EDA and ANOVA ($\alpha=$ 5%) with the 'dose' variable. <br>
Summarize the result.

```python
# Categories of 'dose'. 
df.dose.unique()
```

```python
# Group by category.
x_1= len(df[df.dose==0.5])
x_2 = len(df[df.dose==1.0])
x_3 = len(df[df.dose==2.0])
print(x_1,x_2,x_3)
```

```python
# Multiple Boxplots.
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.boxplot([x_1,x_2,x_3],0)
plt.show()
```

```python
# ANOVA.
res = st.f_oneway(df.dose==0.5,df.dose==1.0,df.dose==2.0)

print("Test statictic  : %f " % res[0] )
print("p-value         : %f " % res[1] )
res
```
