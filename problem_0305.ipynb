### "Correlation and interval estimation"

```python
import numpy as np
import scipy.stats as st
```

#### Answer the following questions by providing Python code:

Calculate the 95% and 99% conficence intervals for the Pearson correlation. <br>
Discuss whether the confidence interval includes zero or not.

1). When the sample size $n = 30$ and the sample correlation $r = 0.45$:

```python
n=30
r=0.45
r_z = np.arctanh(r)
se = 1/np.sqrt(n-3)
```

```python
# 95% confidence interval. 
z = st.norm.ppf(1-0.05/2)
lo_z, hi_z = r_z-z*se, r_z+z*se
lo, hi = np.tanh((lo_z, hi_z))
print (f'The confidence interval is : [', lo, f' , ',hi, f']')
```

Zero isn't included in the interval

```python
# 99% confidence interval. 
z = st.norm.ppf(1-0.01/2)
lo_z, hi_z = r_z-z*se, r_z+z*se
lo, hi = np.tanh((lo_z, hi_z))
print (f'The confidence interval is : [', lo, f' , ',hi, f']')
```

zero is included in the interval
